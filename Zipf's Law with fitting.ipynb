```python
fp=open('Alice.txt','r',encoding='UTF-8')
my_dic={}
for word in fp.read().split():
    if word not in my_dic:
        my_dic[word] = 1
    else:
        my_dic[word] += 1
print (word,my_dic)
fp.close();

import matplotlib.pyplot as plt
num=[]

for key in my_dic:
    num.append(my_dic[key])
    
print(len(num))
num.sort()
num.reverse()

plt.loglog(range(len(num)),num)
plt.show

nums = num[1:]

import numpy as np
logx = np.log(len(num))
logy = np.log(num)

from scipy import optimize

def test_func(x, amp, alpha):
    return amp*x**alpha


params, params_coveriance = optimize.curve_fit(test_func, range(1,len(num)),nums)
print(params, params_coveriance)
plt.loglog(range(1,len(num)),test_func(range(1,len(num)),params[0], params[1]))

plt.show()
```

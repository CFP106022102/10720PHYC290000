```python
%matplotlib inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

im1 = mpimg.imread('20190321_155046.jpg')
im2 = mpimg.imread('20190321_155047.jpg')
im3 = mpimg.imread('20190321_155048.jpg')
im4 = mpimg.imread('20190321_155049.jpg')
im5 = mpimg.imread('20190321_155055.jpg')

b_im1 = im1[:, :, 2]
b_im2 = im2[:, :, 2]
b_im3 = im3[:, :, 2]
b_im4 = im4[:, :, 2]
b_im5 = im5[:, :, 2]

b_im = np.stack((b_im1, b_im2,b_im3, b_im4, b_im5))

print(b_im.shape)
print(b_im)
```

```python
print(np.mean(b_im))
b_ave = np.mean(b_im, axis=0)
plt.imshow(b_ave)
```

```python
im = np.stack((im1, im2,im3, im4, im5),axis=0)

print(im.shape)

im_ave = np.mean(im, axis=0)

print(type(im_ave))

plt.imshow(im_ave.astype(int))
```
